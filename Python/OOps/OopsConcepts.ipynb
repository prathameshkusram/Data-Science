```python
#Function decorator and class decorators

# Function Decorators
# To understand use case >> say you want to use the line before compuation and after computation4
#after each time you create a function or call function so it will take a lot of time to tyoe the lines repeatatively
def my_decorator_func():
    print("The line before computation")
    print(11*1200)# decorating the actual computation with line above and line below
    print("This line after computation")
#In above approch you have to wirte all the line as many times as you are creating the different functions
#lets see the decorator approach
```

```python
my_decorator_func()
```

```python
# Decorator approach for functions>> use case 1
def my_decorator(func): # Decorator function that takes another function as arguement
    def wrapper(): #Wrapper is a nested function that adds the functionality before and after calling original function
        print("The lines before computation")
        func()
        print("This line after computation")
    return wrapper
```

```python
@my_decorator #Syntax to decorate the say hello function
def say_hello():
    print("Hello")
```

```python
say_hello()
#when Say_hello is called,it is actually first call the decorator method >>
# which in return is calling wrapper function and then wrapped function is printing the line and calling the say_hello function
```

```python
#another use case of function decorator
# run time of a code
import time
def timer_decorator(func):
    def timer():
        start=time.time()
        func()
        end=time.time()
        print("The time for executing the code",end-start)
    return timer
```

```python
@timer_decorator
def func_test():
    print(11111*2394932)#any operation
```

```python
func_test()
```

```python
@timer_decorator
def func_test1():
    print(11111+2394932)#any other operation
func_test1()
```
