```python
#Numerical data plotting
# relplot has options of plotting both categorical and numerical data
# relplot >> show the relationship between two variable
# relplot is relationship plot
df
```

```python
# how tip is varying wrt total_bill>>Both are numeric >> Scatter >>by default relplot makes scatter plot

sns.relplot(x='total_bill',y='tip',data=df)# relationship between two variable ,by default it givrs scatter plot,if you use kind='line'it will change to line graph
# insight >> as total bil increase tip increases
```

```python
#hue
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker')
#insight >> Customer who smoked gave the highest tip
#hue >> Hue gives the different color shades
```

```python
df
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='size')
#In this plot we can not clearly see the difference >> for this we can use style
```

```python
#Style>> style give the different representation of column value
sns.relplot(x='total_bill',y='tip',data=df,style='size')
```

```python
# now using style and hue in one
#hue --> for different color codes
#style-->for different shape repersention

sns.relplot(x='total_bill',y='tip',data=df,style='smoker',hue='time')
```

```python
df1=sns.load_dataset('fmri')
df1
```

```python
sns.relplot(x='timepoint',y='signal',data=df1,kind='line')
```

```python
sns.relplot(x='timepoint',y='signal',data=df1,kind='line',ci='sd')
#ci='sd' zoom the confidence
# the shaded portion is confidence it is the area up to that it can have value
```

```python
# if you dont want this confidnce
# estimator
sns.relplot(x='timepoint',y='signal',estimator=None,kind='line',data=df1,ci='sd')
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=df1,hue='event')
```

```python
#remove the dashesh in line chart-->dashes=flase
sns.relplot(x='timepoint',y='signal',data=df1,kind='line',hue='event',style='region')
```

```python
#Remove the dashes in line chart >> dashes=False
sns.relplot(x='timepoint',y='signal',data=df1,kind='line',hue='event',style='region',dashes=False)
```
