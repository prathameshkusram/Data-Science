```python
#  Graph objects
# express
import plotly.graph_objects as go
import plotly.express as px
```

```python
# using graph objects
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[1,7,8,2,6],mode='markers'))
fig.show()
```

```python
# line plot
# mode='lines'
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[1,7,8,2,6],mode='lines'))
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers'))
```

```python
#bar chart
fig=go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5],y=[4,3,6,5,2]))
```

```python
fig=go.Figure()
fig.add_trace(go.Histogram(x=tips['total_bill']))
```

```python
# Scater plot
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers'))
```

```python
tips
```

```python
#marker size based of tip size if tip is more the marker is big
#using marker size for increasing the size of marker  you can multiply with integer so that it is clearly visible
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers',marker_size=2*tips['size']))
```

```python
# 3-d Plot
# use go.Scatter3d
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill,y=tips.tip,mode='markers',marker_size=3*tips['size'],z=tips['size']))
```

```python
# 3-D line chart
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=[1,2,3,4,5],y=[3,8,5,2,6],z=[1,2,7,3,5],mode='lines'))
```

```python
# pLotly express px
x=[1,2,3,4]
y=[10,13,15,18]
fig=px.scatter(x,y)
fig.show()
```

```python
# Line plot using expree
fig=px.line(x,y)
fig.show()
```

```python
x=['A','B','C']
y=[10,5,6]
fig=px.bar(x=x,y=y)
fig.show()
```

```python
# histogram using express px
x=[5,10,3,4,5,5,6,6,10,10,2,11,11,12,3,4]
fig=px.histogram(x)
fig.show()
```

```python
# box plot
x=['A','B','C']
y=[5,10,3,25,12,33,100]
fig=px.box(x,y)
fig.show()
```

```python
data=[[1,2,3],[4,5,6],[7,8,9]]
fig=px.imshow(data)
fig.show()
```

```python
# bubble chart
x=[1,2,3,4]
y=[5,8,12,3]
fig=px.scatter(x,y,size=[30,10,50,100])
fig.show()
```

```python
# 3d chart
x=[1,2,3,4]
y=[5,8,12,3]
z=[3,4,5,6]
fig=px.scatter_3d(x=x,y=y,z=z)
fig.show()
```

```python
# Contour plot
x=[1,2,3,4,3,23,80]
y=[5,8,12,3,12,35,60]
fig=px.density_contour(x,y)
fig.show()
```
