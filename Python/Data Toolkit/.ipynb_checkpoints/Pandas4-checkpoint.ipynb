```python
#Merging the dataframes
df1 = pd.DataFrame({
    'key1': [1, 2, 3, 4, 5],
    'key2': [4, 5, 6, 7, 8],
    'key3': [3, 4, 5, 6, 6]
})
df1
```

```python
df2 =pd.DataFrame({
    'key1': [1, 2, 45, 6, 67],
    'key5': [56, 5, 6, 7, 8],
    'key6': [3, 56, 5, 6, 6]
})
```

```python
df2
```

```python
#Inner merge
pd.merge(df1,df2,how='inner')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer')# all the common element of the common column from bhot tho dataframe
```

```python
pd.merge(df1,df2,how='cross')
```

```python
# merging without diffrent name
pd.merge(df1,df2,how='left',left_on='key2',right_on='key5')
```

```python
#join--> join works based on index/merge works based on columns
```

```python
df1 = pd.DataFrame({
    'key1': [1, 2, 4, 5, 6],
    'key2': [4, 5, 6, 7, 8],
    'key3': [3, 4, 5, 6, 6]},
    index=['a','b','c','d','e']
)
df1
```

```python
df2 =pd.DataFrame({
    'key6': [1, 2, 45, 6, 67],
    'key4': [56, 5, 6, 7, 8],
    'key5': [3, 56, 5, 6, 6]},
    index=['a','b','h','i','j']
)
df2
```

```python
df1.join(df2,how='inner')#joins columns of another dataframe based on common index
```

```python
df1.join(df2,how='left')
```

```python
df1.join(df2,how='right')
```

```python
df1.join(df2,how='outer')
```

```python
df1.join(df2,how='cross')
```

```python
def create_flag(x):
    if x <10:
        return "cheap"
    if x>=10 and x<20:
        return 'medium'
    else:
        return 'high'
```

```python
df['Flag_fare']=df['Fare'].apply(create_flag)
```
