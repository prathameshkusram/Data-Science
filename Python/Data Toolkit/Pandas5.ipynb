```python
# to check the execution /run time of a code
start=pd.Timestamp.now()
for i in range(1000):
    pass
end=pd.Timestamp.now()
print('Time taken for the execution of code',end-start)
```

```python
# Another way
import time
start=time.time()
for i in range (1000):
    pass
end=time.time()
print('Time taken for the execution of code in second ',end-start)
```

```python
#One more way to check execution time
import time
start=time.perf_counter()
for i in range(1000):
    pass
end=time.perf_counter()
print('Time required for the execution of code in second',end-start)
```
