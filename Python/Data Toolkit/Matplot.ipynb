```python
#Line plot
#Whenever you want to depict the relation between two variables in the form of trend
# Mostly in x axis the date - time column is there

x=[1,2,3,4,5]# assume x is no days
y=[5,2,7,8,1]#price of stock
plt.plot(x,y)
```

```python
plt.plot(x,y,)
plt.title('Line chart to understand the syntax')
plt.xlabel("Day interval as x points")
plt.ylabel("Price as y points")
```

```python
plt.plot(x,y,color='red')
plt.title('This is line chart')
plt.xlabel("Day interval as x point")
plt.ylabel("Price as y point")
```

```python
plt.plot(x,y,color='red',marker='o')
plt.title('This is line chart')
plt.xlabel("Day interval as x point")
plt.ylabel("Price as y point")
```

```python
plt.plot(x,y,color='red',marker='o',linestyle='--')
plt.title('This is line chart')
plt.xlabel("Day interval as x point")
plt.ylabel("Price as y point")
plt.show()
```

```python
plt.plot(x,y,color='red',marker='o',linestyle='--',linewidth=3)
plt.title('This is line chart')
plt.xlabel("Day interval as x point")
plt.ylabel("Price as y point")
plt.show()
```

```python
plt.plot(x,y,color='red',marker='o',linestyle='--',linewidth=3,markersize=10)
plt.title('This is line chart')
plt.xlabel("Day interval as x point")
plt.ylabel("Price as y point")
plt.show()
```

```python
x=np.linspace(0,10,100)
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,np.sin(x),linestyle=':')
plt.plot(x,np.cos(x))
```

```python
# how to make two plots in on go
plt.figure()# Initialising a page to make plot
plt.subplot(1,2,1)#(1 -> 1 row,2 --> 2 column,1-> 1st graph)
plt.plot(x,np.sin(x),color='red')

plt.subplot(1,2,2)#(1 -> 1 row,2 --> 2 column,2-> 2st graph)
plt.plot(x,np.cos(x))

plt.show()#Display all open figures.
```

```python
# how to make two plots in on go
plt.figure()# Initialising a page to make plot
plt.subplot(2,1,1)#(2-> 2 row,1 --> 1 column,1-> 1st graph)
plt.plot(x,np.sin(x),color='red')

plt.subplot(2,1,2)#(2 -> 2 row,1 --> 1 column,2-> 2st graph)
plt.plot(x,np.cos(x))

plt.show()#Display all open figures.
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

plt.subplot(2,2,3)
plt.plot(x,np.tan(x))

plt.subplot(2,2,4)
plt.plot(x,np.sin(x+1))
```

```python
plt.plot(x,np.sin(x),color='red')
```

```python
# usning limit
plt.plot(x,np.sin(x-0),color='red')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.show()
```

```python
#using Axis make graph to cover the remaining part
plt.plot(x,np.sin(x-0),color='red')
plt.ylim(-1.5,1.5)
plt.xlim(1,15)
plt.axis('Equal')
plt.show()
```

```python
# Bar plot >> to see the districution of categorical variables
x=["a","b","c","d"]#Name of students
y=[4,6,1,10]# No of days present in the class
plt.bar(x,y,color='r')
```

```python
# pie chart >> use to understand the percent distribution of data
langs=['c','Python','Java']
students=[20,30,67]
plt.pie(students,labels=langs,autopct='%1.1f%%')
plt.title('Pie chart')
plt.show()
```

```python
#Explode
explode=(0.0,0.1,0.1)
langs=['c','Python','Java']
students=[20,30,50]
plt.pie(students,labels=langs,autopct='%1.1f%%',explode=explode)
plt.title('Pie chart')
plt.show()
```

```python
#Shadow
explode=(0.0,0.1,0.1)
langs=['c','Python','Java']
students=[20,30,50]
plt.pie(students,labels=langs,autopct='%1.1f%%',explode=explode,shadow=True)
plt.title('Pie chart')
plt.show()
```

```python
year=[2021,2018,2022,2023,2017]
category1=[30,25,38,28,35]
category2=[10,15,20,18,25]
category3=[5,8,10,9,15]
plt.stackplot(year,category1,category2,category3,labels=['category1','category2','category3'],
             colors=['skyblue','lightgreen','orange'])
plt.xlabel('years')
plt.ylabel('Values')
plt.title('Area plot')
plt.show()
```
