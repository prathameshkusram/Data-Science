```python
# Re-shaping of array
import random
newarr=np.random.randint(1,10,size=(4,3))
newarr
```

```python
newarr.size
```

```python
#using reshape 
newarr.reshape(2,6)#Returns an array containing the same data with a new shape.
```

```python
# Array size will never change,always reshape parameter should be multiple of original array
newarr.reshape(3,3)
# show an error because the original array size is 12
```

```python
# In case you dont know coloums  in reshape what should be the row or coloumn count 
# use any -ve value for rows or coloums
newarr.reshape(4,-5)
```

```python
#same for rows
newarr.reshape(-1,4)
```

```python
newarr.reshape(4,3).base# to see the original array
```

```python
#3d
newarr.reshape(2,3,2)  #multiplication should be equal to arr size
```

```python
#4d
newarr.reshape(3,2,2,1)
```

```python
newarr.reshape(3,2,2,1).ndim
```

```python
newarr.reshape(4,3,1)# 3 rows 1 coloumn repeated 4 times in 3-dimensiion
```
