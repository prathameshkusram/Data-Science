```python
# What is exception?
#--> Generally something that disrups the normal event is exception

# In Python an exception is an event that occurs during programe execution that disrupts the regular flow of code
```

```python
# you were expecting something but it didnt happen,due to which the suspicious line threw error and the next line of code didn't got executed
# Zero divison error,type error,file not found error -->these are exceptions rising due to suspious code
# other exception >> import error ,runtime error

# why do we need to handle this suspicious code/exception?
#--> Robust application
#--> Code cleaner --> fellow developer understands the corner cases


# how to handle exception
#--> using try catch block to handleexception
#--> whenever exception occurs >> python stops the program and generate exception
#--> handling this exception is advised
#--> every code that is risky/suspicious you should do exception handling
```

```python
# try except method

#try :
#    suspicious code
#except : 
#   executed when exception occurs in try block
```

```python
try:
    f=open("bye_bye.txt",'r')
except Exception as e:
    print("There is an issue with code and the issue is ",e)
```

```python
try:
    5/0
except Exception as e: #Exception will always come inside except block
    print("The error is ",e)
```

```python
try:
    f=open("example.txt",'r')
except Exception as e:
    print("There is an issue with code and the issue is ",e)
else:
    print("This will be executed if try except block is executed without any exception")
```

```python
# use case of else>> you wantt to execute some code only of the try statement is successful and code does'nt enter the exception block
```

```python
try:
    f=open("example.txt",'w')
    f.write("Hello to exception handling")# you wil not be able to see written content until you close the file
    # close the file in else block
except Exception as e:
    print("There is an issue with code and the issue is ",e)
else:
    f.close()
    print("This will be executed if try except block is executed without any exception")
```

```python
# else block is very helpful when your  code didnt enter the exception block
# for example you wanted to close the file after it was opened in try block
```

```python
# Finally Block >> Always executed
# weather if you enter exception block or not finally will be exceuted always
# similar to running other lines of code
try:
    f=open("example.txt",'w')
    f.write("Hello to exception handling")
except Exception as e:
    print("There is an issue with code and the issue is ",e)
else :
    f.close()
    print("This will be executed if try except block is executed without any exception")
finally:
    print("This will be executed always")
```

```python
# another example
try:
    5/0
except Exception as e:
    print(e)
finally:
    print("This will be printed always")
```

```python
# Custom exception class
class validatesalary(Exception): # Exception is parent class where all type of python exception are available
    def __init__(self,msg):
        self.msg=msg
```

```python
def validate_salary(salary):
    if salary<=0:
        raise validatesalary("Negative salary is not possible")
    elif salary >=3000000:
        raise validatesalary("Salary is unexpected")
    else:
        print("Salary is valid")
        
```

```python
try:
    salary=int(input("Enter your salary : "))
    validate_salary(salary)
except Exception as e:
    print(e)
```
