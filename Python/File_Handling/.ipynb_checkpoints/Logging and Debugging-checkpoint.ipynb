```python
def divide(a,b):
    return a/b
for i in range(-10,10):
    divide(i,i-1)
# Breaking  the code>> Logicaly>>Here we understand that there are two steps involved
# Break the code in terms of exceution as will using return statements or print statements
#first a loop second callinh divide function in loop 
```

```python
# using print statement to debug
def divide(a,b):
    return a/b
    
for i in range(-10,10):
    print(i,i+1)
    divide(i,i+1)
```

```python
#Using another print statement inside function
def divide(a,b):
    print(a,b)
    return a/b
    
for i in range(-10,10):
    #print(i,i+1)
    divide(i,i+1)
# Form this we conclude that whenever the b is 0 the code is throwing an error
```

```python
# from print statement you understand that when b is 0 ,it throws the error
def divide(a,b):
    print(a,b)
    # breaking the code in terms of execution
    if b==0:
        print("Zero divison error")
        return
    return a/b
    
for i in range(-10,10):
    #print(i,i+1)
    divide(i,i+1)
```
