```python
# Best Practice 1: Always catch specific error
# specificity in exception handling
try: 
    10/0 # suspicious 
except ZeroDivisionError as e:
    print("Here I am handling the zero divison error",e)
```

```python
#Another example
try :
    10/"some string"
except TypeError as e:
    print("Some string is begin used in divison",e)
```

```python
# Benefit of handling specific error>> you know what error is expected and if other error comes which you were not able to think ,you be able to address that
```

```python
# Example to handle some specific exception
try:
    10/"3"
except TypeError as e:
    print("Some string is being used in divison",e)
    print("The calculation is",10/int("3"))
    
```

```python
# another example of handling specific exception
try:
    10/0
except ZeroDivisionError as e:
    print("Here I am handling the zero divison error",e)
    print("Since the denominator is 0,using default value 1,the quotient is",10/1)
```

```python
# Handling two specific exception
# 1]
try:
    10/0
except (ZeroDivisionError,TypeError) as e:
    print("The divison is not possible due to the error",e)
```

```python
# 2]
try:
    10/"2"
except (ZeroDivisionError,TypeError) as e:
    print("The divison is not possible due to error",e)
```

```python
# Handling multiple exception in separate block
try:
    10/"4"
except ZeroDivisionError as e:
    print("The divison is not possible due to the error",e)
except TypeError as e:
    print("This is value error",e)
```

```python
# sometimes you will not know what could be the possible exception
try:
    f=open("Pandu.txt",'r')
    f.read()
except Exception as e: # Parent class
    print("File is not opend due to",e)
except FileNotFoundError as e:
    print("The following file was not created ,please try again later",e)
```

```python
# in case you know specific exception keep that block first and the block of generic exception after that
try:
    f=open("Pandu.txt",'r')
    f.read()
except FileNotFoundError as e:#parent class
    print("The following file was not created ,please try again later",e)
except Exception as e: 
    print("File is not opend due to",e)
```

```python
# best practice >> Always try to log

import logging
logging.basicConfig(filename="Program.log",level=logging.DEBUG)
try:
    f=open("Pandu.txt",'r')
    f.read()
except FileNotFoundError as e:
    logging.error(f"My file was not found{e}")
```

```python
# putting exception handling in a function 
def test(file):
    try:
        f=open(file,"r")
        f.read()
    except FileNotFoundError as e:
        print(f"My file was not found {e}")
```

```python
test("pandu.txt")
```
