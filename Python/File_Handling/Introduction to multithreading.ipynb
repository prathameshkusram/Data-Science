```python
# iF we call same function multiple times
import time
import threading # python module
start = time.perf_counter()

def test_func():
    print("Do something")
    print("Sleep for 1 sec")
    time.sleep(1)
    print("Done with sleeping")
    
# Run the program on two threads
t1=threading.Thread(target=test_func)
t2=threading.Thread(target=test_func)

t1.start()# to start thread
t2.start()

t1.join()# join first executed these t1 ,t2 threads and then the main thread will be executed
t2.join
end=time.perf_counter()

print(f"The program  finished in {round(end-start,2)} seconds")
```

```python
# iF we call same function multiple times
import time
import threading # python module
start = time.perf_counter()

def test_func():
    print("Do something")
    print("Sleep for 1 sec")
    time.sleep(1)
    print("Done with sleeping")
    
# Run the program on two threads
threads=[]
for i in range(10):
    t=threading.Thread(target=test_func)
    t.start()
    threads.append(t)

for thread in threads:
    thread.join()


end=time.perf_counter()

print(f"The program  finished in {round(end-start,2)} seconds")
```

```python
# since test_func is called in 10 times,it should have taken 10 secs but due to multithreading got completed in 1 sec
```

```python
#https://github con/itsfoss/t files
#Downloading files form git hub

import time
import threading
start=time.perf_counter()
url_list=['https://raw.githubusercontent.com/itsfoss/text-script-files/refs/heads/master/agatha_complete.txt',
          'https://raw.githubusercontent.com/itsfoss/text-script-files/refs/heads/master/sample_log_file.txt',
          'https://raw.githubusercontent.com/itsfoss/text-script-files/refs/heads/master/sherlock.txt',
          'https://raw.githubusercontent.com/itsfoss/text-script-files/refs/heads/master/agatha.txt'
         ]

data_list=["data1.txt","data2.txt","data3.txt","data4.txt"]

import urllib.request
def file_downlod(url,filename):
    urllib.request.urlretrieve(url,filename)

threads=[]
for i in range (len(url_list)):
    t=threading.Thread(target=file_downlod,args=(url_list[i],data_list[i]))
    t.start()
    threads.append(t)

for thread in threads:
    thread.join()

end=time.perf_counter()
print(f"The program finished in {round(end-start,2)} seconds")
```

```python
#Multithreading using concurrent futures >> keep code concise
#https://github con/itsfoss/t files
#Downloading files form git hub

import time
import concurrent.futures
start=time.perf_counter()
url_list=['https://raw.githubusercontent.com/itsfoss/text-script-files/refs/heads/master/agatha_complete.txt',
          'https://raw.githubusercontent.com/itsfoss/text-script-files/refs/heads/master/sample_log_file.txt',
          'https://raw.githubusercontent.com/itsfoss/text-script-files/refs/heads/master/sherlock.txt',
          'https://raw.githubusercontent.com/itsfoss/text-script-files/refs/heads/master/agatha.txt'
         ]

data_list=["data1.txt","data2.txt","data3.txt","data4.txt"]

import urllib.request
def file_downlod(url,filename):
    urllib.request.urlretrieve(url,filename)

    with concurrent.futures.ThreadPoolExecutor() as executor:
        executor.map(file_downlod,url_list,data_list)#args>>function name with arguments

end=time.perf_counter()
print(f"The program finished in {round(end-start,2)} seconds")
```

```python
#Shared variable  across all the threads
start=time.perf_counter()
shared_counter=0
counter_lock=threading.Lock()#Locking the counter for a specific thread

def increment_shared_counter(x):
    global shared_counter#That can be accessed by all the threads
    with counter_lock:
        shared_counter=shared_counter+1
        print(f"Thread{x}:incremented shared counter to {shared_counter}")
        time.sleep(1)
threads=[threading.Thread(target=increment_shared_counter,args=(i,)) for i in[1,2,3,4,5,6]]

for thread in threads:
    thread.start()
for thread in threads:
    thread.join()
end=time.perf_counter()
print(f"The program finished in{round(end-start,2)} seconds")
```

```python
# same thing using concurrent futures
#Shared variable  across all the threads
start=time.perf_counter()
shared_counter=0
counter_lock=threading.Lock()#Locking the counter for a specific thread

def increment_shared_counter(x):
    global shared_counter#That can be accessed by all the threads
    with counter_lock:
        shared_counter=shared_counter+1
        print(f"Thread{x}:incremented shared counter to {shared_counter}")
        time.sleep(1)
with concurrent.futures.ThreadPoolExecutor() as executor:
        thread_args=[1,2,3,4,5,6]
        executor.map(increment_shared_counter,thread_args)#args>>function name with arguments
    
end=time.perf_counter()
print(f"The program finished in{round(end-start,2)} seconds")
```
