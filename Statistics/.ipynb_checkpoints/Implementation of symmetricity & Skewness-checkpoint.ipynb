```python
data=[2,3,4,5,8,9,12,40,41,15]
data
```

```python
# Meassure of dispersion/Range
# How too calculate range
#1]
max(data)-min(data)
```

```python
# 2] Percentlie
import numpy as np
np.percentile(data,[0,25,50,75,100])
```

```python
# From above we conculde that 50 percent of data is below 8.5
```

```python
# 3] IQR
14.25-4.25
```

```python
#4]  Box plot
import seaborn as sns
sns.boxplot(data)# upper fence>>Q3+1.5*IQR ,Lower fence = Q1-1.5*IQR
```

```python
# Variance
np.var(data)
```

```python
# Standard Deviation
np.std(data)
```

```python
# Another way of calculating  variance and standar deviation
```

```python
import statistics
statistics.variance(data)# this is sample variance>> In denominator it is divided by n-1
```

```python
statistics.pvariance(data)# pvariance will give you population variance
```

```python
# Measure of symmetricity ,Skweness
# Skewness

from scipy.stats import  skew
skew(data)# Since skewness value is 1.23 it is moderately skewed data
```

```python
d=[2,3,4,5,6,8,11]
skew(d)# almost same range so skewness close to 0
```
